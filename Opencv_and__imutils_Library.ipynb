```python
#import required libraries
import cv2
import imutils
```

```python
#read the orginal image
real=cv2.imread('/content/mahesh-babu_153379369590.jpg')
```

```python
real=imutils.resize(real,width=10000)
#convert real to gray image
grayimg=cv2.cvtColor(real,cv2.COLOR_BGR2GRAY)
#CREATE THRESHOLD IMAGE
threshimg=cv2.threshold(grayimg,200,255,cv2.THRESH_BINARY)[1]
canny=cv2.Canny(real,500,75)
#removw the blank spots
dialted=cv2.dilate(canny,(7,7))
```

```python
# @title Default title text
#save the converted images
cv2.imwrite("grayimg.jpg",grayimg)
cv2.imwrite("threshimg.jpg",threshimg)
```
