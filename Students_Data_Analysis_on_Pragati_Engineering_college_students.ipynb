# Find the average commute_Time using Sampling method

```python
#taking sample size of 30
sample_size=30
#Creating multiple number of samples and calculating their means
sample_mean=[df["Commute_Time"].sample(sample_size).mean() for i in range(1000)]
```

```python
#sample means
sample_mean
```

```python
#storing sample mean values in a series using pandas
sample_mean=pd.Series(sample_mean)
```

```python
#verifying the total number of samples collected
len(sample_mean)
```

```python
#plotting the density for the sample means
sns.histplot(x=sample_mean,kde=True)
```

**Observation:*


*  As  we can observe that above distribution looks like Normal Distribution



```python
#Mean value of all then 1000 samples
sample_mean.mean()
```

```python
#Standard deviation of all the 1000 samples
sample_mean.std()
```

**Observation:**

1.   sample mean =85.357966 which is almost similar to Original mean value 85.0 and sample std=7.3719931372076895


#Finding Confidence Intervel



1.   Noe we have the mean of commute time of 72 employes is 85 and standard deviation of the sample is S=7.3719931372076895 min
2.   Using Central limit theorem we can infer that




**1.samplig distribution = population distribution**

**Sampling distribution of standard deviation=52.12640561329893/sqr(72)=6.14315581467**

**the mean of commute time of 30000 students is 85+Margin of error**



*   Because this not enough to say all the students commute time is 85 min by considering only 72 students in college



**Formula for margine of error**

***Margin of error = Critical value * Standard Error***

**For 95% confidence level the critical value is 1.96**

**standard deviation of the sample is called Standard error(7.3719931372076895)**

```python
critical_value=1.96
standard_error=7.3719931372076895
```

```python
Error=critical_value*standard_error
```

**Confidence Intervel=Sample Estimate +/- Margin of Error**

```python
sample_mean.mean()+Error
```

```python
sample_mean.mean()-Error
```

```python
print(f'I am 95% Confidence that PRAGATI ENGINEERING COLLEGE student commute time is lie between {sample_mean.mean()-Error} and {sample_mean.mean()+Error} minutes')
```
